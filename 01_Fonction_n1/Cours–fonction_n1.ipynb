## Introduction
Une fonction est un morceau de code isol du reste du programme, qui possde un *nom* et qui peut tre appel par ce nom  n'importe quel endroit du programme, autant de fois que l'on veut. Une fonction peut tre vue comme une sorte d'usine qui a besoin de matire premire pour fonctionner et produire des objets. Faisons le parallle entre une usine d'embouteillage et une fonction.

## Dfinition d'une fonction
Dans le langage Python, une fonction est dclare par l'instruction `def` (pour definition), suivie du nom donn  la fonction, et d'une paire de parenthses qui contient les **arguments** ou **paramtres** ventuellement passs  la fonction lors de son appel. Tout le contenu de la fonction est **indent**. Voici un exemple :

````python
def traceUneLigne40():
	""" Cette fonction trace une ligne de 40 tirets.
	Argument(s) : aucun
	Valeur de retour : aucune """
	print("-" * 40)
````

La cellule, ci-dessus, donne la dfinition de la fonction `traceUneLigne40` mais n'excute pas le code. L'excution du code a lieu lorsque l'on appelle la fonction :

````python
traceUneLigne40()
````

Pour pouvoir utiliser une fonction dans un programme, celle-ci doit avoir t dfinie en amont de son appel.   
Dans la dfinition de la fonction, une rapide explication de son fonctionnement est donne dans le commentaire dlimit par `"""` (qui permet d'crire un commentaire sur plusieurs lignes). Ce commentaire permet de faciliter la lecture du programme aux autres dveloppeurs. On appelle cette partie *docstring*. La commande ```help(nomFonction)```, insre dans votre programme, provoquera son affichage.

````python
help(traceUneLigne40)
````

Enfin, dans la dfinition de la fonction, on trouve bien videmment le code  excuter : `print("-" * 40)`.

### Arguments
Comme dans une recette de cuisine o des ingrdients sont ncessaires pour faire un gteau, une fonction a souvent besoin d'information supplmentaires pour raliser son calcul. Ces  informations supplmentaires constituent les **_arguments_** (paramtres d'entres) de la fonction que l'on prcise  la suite de son nom :
```python
def crepe(poids_farine, nb_oeufs, volume_lait):
```
L'utilisation des _annotation de type_ se gnralise depuis la version 3.5 de Python. Il s'agit d'indiquer le type de paramtres (`poids_farine : float`) et le type de valeur de retour (`-> int`) si la fonction renvoie le nombre de crpes prpares. Ces annotations sont toutefois facultatives.

Dans la fonction dfinie plus haut, supposons maintenant que nous voulions rendre paramtrable facilement le nombre de tirets  afficher :

````python
# Dfinitions
def traceUneLigne(nbTirets:int):
    """ Cette fonction trace une ligne de 40 tirets.
    Argument(s) : nbTirets (int) 
    Valeur de retour : aucune """
    print("-" * nbTirets)

    
# Programme principal
traceUneLigne40()        # fonction du dbut
traceUneLigne(20)        # Nouvelle fonction avec un paramtre
````

Nous avons ajout en argument de la fonction, la variable `nbTirets` qui est utilis directement dans la fonction `print()` pour dfinir le nombre de tirets  afficher.

### Porte des variables de la fonction
Les variables dfinies dans une fonction **n'existent pas  l'extrieur** de celle-ci. On dit qu'elles ont une __porte locale__.  
Vous pouvez le vrifier avec l'excution du programme ci-dessous o une fonction qui calcule la puissance d'un nombre a t cre. Le programme essaie ensuite d'utiliser la variable `resultat` dfinie dans la fonction dans le programme principal.

````python
# Dfinition de fonctions
def puissance(a:float,b:int)->float:

    resultat = a ** b

# Programme principal
x = 3.0                 #Nombre que je veux lever  la puissance 3.   
puissance(x,3)          # J'utilise ma fonction
print("Rsultat :", resultat)   # J'essaie d'afficher la variable resultat utilise plus haut...
````

 la variable `resultat` ne peut donc pas tre utilise en dehors de la fonction ! 

_Mais comment faire pour rcuprer la rsultat du calcul d'une fonction ?_

### Valeur en retour
Pour pouvoir accder au rsultat d'un calcul effectu dans une fonction, il faut forcer la fonction  **retourner** une valeur  l'extrieur de celle-ci  l'aide du mot cl `return`.  noter que cette valeur peut trs bien tre le contenu d'une variable.

````python
# Dfinition de fonctions
def puissance(a:float,b:int)->float:

    resultat = a ** b
    return(resultat)    #la fonction retourne la valeur contenue dans la variable resultat

# Programme principal
x = 3.0                  #Nombre que je veux lever  la puissance 3.   
res = puissance(x,3)     # grce au return nous pouvons rcuprer le rsultat du calcul et le stocker dans une variable
print("Rsultat :", res)
````

 Une fois la valeur retourne par la fonction, il faut la **rcuprer** dans le programme principal et la stocke dans une **variable** comme dans l'exemple prcdent avec la variable `res`.
